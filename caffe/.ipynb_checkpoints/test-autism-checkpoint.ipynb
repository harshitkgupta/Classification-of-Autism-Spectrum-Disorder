```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from PIL import Image as Image
import skimage.io
%matplotlib inline
```

```python
# Make sure that caffe is on the python path:
# this file is expected to be in {caffe_root}/examples/speech
caffe_root = '/home/mmlab/Downloads/caffe-master/' 

sys.path.insert(0, caffe_root + 'python')
sys.path.insert(0, caffe_root + 'python/caffe')

import caffe

model="numbers_deploy.prototxt"
weights="snapshot_iter_18.caffemodel"

net = caffe.Net(model,weights,caffe.TEST)
caffe.set_mode_cpu()





# some reflections:
#help(net)
print net.params
print net.layers #LayerVec
print net.layers[0] #PyLayer __reduce__() .blobs
#c++: layer_param,loss(int top_index),type_name(),
#print net.layers[0].blobs[0] #Blob 
# Blob   channels, count,data,diff,height,num,width
#print net.layers[0].type_name #added in branch
```

```python
net = caffe.Classifier(model, weights,
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy').mean(1).mean(1),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(512,512))


for i in range(1,5):
    image_file='ASD/f'+str(i)+'.png'
    input_image = caffe.io.load_image(image_file)
    prediction = net.predict([input_image])  # predict takes any number of images, and formats them for the Caffe net automatically
    print 'prediction shape:', prediction.shape
   # plt.plot(prediction[0])
    print 'predicted class:', prediction[0].argmax()
    #plt.show()
```
